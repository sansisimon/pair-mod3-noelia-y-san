```python
def limpieza (texto):
    reemplazo = str(texto).replace(",", "")
    return reemplazo
```

```python
df1["land_area"].apply(limpieza)
```

```python
df1["land_area"].info()
```

```python
lista_columnas = ["land_area", "armed_forces_size", "co2-emissions", "gdp", "population", "urban_population", "latitude", "longitude", "density", "cpi"]
```

```python
for col in lista_columnas:
    df1[col] = df1[col].apply(limpieza)
```
